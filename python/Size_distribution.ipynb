```python
df = xr.open_dataset("./bin/output.nc")
df
```

```python
Nbin = 40
Nt = len(df['time'])
Nid = len(df['expt_id'])
Aerosol = np.zeros([Nbin,Nt,Nid])

for Id in range(1,Nbin+1,1):
    VarName = f'spc_bin{Id:03d}'
    Aerosol[Id-1,:,:] = df[VarName]*df['dens_air']
```

```python
radius = []

file1 = open('./bin/Radius_Bin_um.txt', 'r')
Lines = file1.readlines()
 
# Strips the newline character
for line in Lines:
    radius.append(float(line))

radius = np.array(radius)
log_r = np.log(radius)
print(len(radius))
```

```python
# r_edge
r_edge = []
r_edge.append(radius[0]-0.5*(radius[1]-radius[0]))
for i in range(1,Nbin,1):
#     print(i)
    r_edge.append(0.5*(radius[i]+radius[i-1]))
r_edge.append(radius[Nbin-1]+0.5*(radius[Nbin-1]-radius[Nbin-2]))


# Dr
Dr = []
for i in range(1,Nbin+1,1):
    Dr.append( r_edge[i]-r_edge[i-1] )


# Dlog_r
Dlog_r = []
for i in range(1,Nbin+1,1):
    Dlog_r.append( np.log(r_edge[i])-np.log(r_edge[i-1]) )
# print(Dlog_r)
```

```python
fig = plt.figure(dpi=100)

plt.plot(radius, Aerosol[:,1,0],'g-', label='1 min')
plt.plot(radius, Aerosol[:,61,0],'c-', label='1 hour')
plt.plot(radius, Aerosol[:,-1,0],'r-', label='1 day')

plt.xscale("log")
plt.yscale("log")

plt.ylim(1e-2,1e9)
# plt.xlim(4e-4,1e0)


plt.xlabel('Aerosol radius [um]')
plt.ylabel('Aerosol number concentration [#/cm3]');
plt.legend();
```

```python
plt.plot(radius, Aerosol[:,1,0]/Dr,'-')
plt.plot(radius, Aerosol[:,61,0]/Dr,'-')
plt.plot(radius, Aerosol[:,-1,0]/Dr,'-')

plt.xscale("log")
plt.yscale("log")

plt.ylim(1e-2,1e12)
plt.xlim(4e-4,1e0)


plt.xlabel('Aerosol radius [um]')
plt.ylabel('dN/dr');
```

```python
fig = plt.figure(dpi=100)

plt.plot(radius, Aerosol[:,1,0]/Dlog_r,'g-', label='1 min')
plt.plot(radius, Aerosol[:,61,0]/Dlog_r,'c-', label='1 hour')
plt.plot(radius, Aerosol[:,-1,0]/Dlog_r,'r-', label='1 day')

plt.xscale("log")
plt.yscale("log")

plt.ylim(1e-2,1e10)
plt.xlim(3e-4,1e0)


plt.xlabel('Aerosol radius [um]')
plt.ylabel('dN / dlog r');
plt.legend();
```
